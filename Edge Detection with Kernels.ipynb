```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
%matplotlib inline

#Read the image 
image = mpimg.imread('ellen-degeneres.jpg')
plt.imshow(image)
```

```python
#Convert the image to grayscale for filtering
grayScaleImg = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(grayScaleImg,cmap='gray')
```

```python
#We will use Sobel filter, commonly used in edge detection in finding
#patterns in intensity in an image
#Applying a Sobel filter to an image is a way of taking (an approximation) of the derivative of the image in the x or y direction, separately.
#The operators look as follows.

#edge detection in 
sobel_y = np.array([[-1,-2,-1],
                    [0,0,0],
                    [1,2,1]])

# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image = cv2.filter2D(grayScaleImg, -1, sobel_y)
plt.imshow(filtered_image,cmap='gray')
```

sobel_x = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])
filtered_image = cv2.filter2D(grayScaleImg, -1, sobel_x)
plt.imshow(filtered_image,cmap='gray')

```python
sobel_decimal = np.array([[0.5,0.3,0.1],[0.1,0.2,0.1],[0.4,0.3,0.4]])
filtered_image = cv2.filter2D(grayScaleImg, -1, sobel_decimal)
plt.imshow(filtered_image,cmap='gray')
```
